```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "alldata.csv"
jobs_data = pd.read_csv(csv_file)
jobs_data.head()
```

```python
jobs_data= jobs_data[jobs_data.location == "Austin, TX"]
jobs_data
```

```python
jobs_data = jobs_data[["position", "company", "location"]]
jobs_data = jobs_data.reset_index()
del jobs_data["index"]
jobs_data.head()
```

```python
jobs_data['company'].value_counts()
```

```python
jobs_data.describe()
```

```python
connection_string = "postgres:changeme@localhost:5432/jobs_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
jobs_data.to_sql(name='jobs_table', con=engine, if_exists='append', index=False)
```
